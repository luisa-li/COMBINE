```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
rms = pd.read_csv("02_03.rms", sep="       ")
rms.reset_index(inplace=True)
rms.columns = ["Frame", "RMSD"]
rms
```

```python
plt.figure(figsize=(20, 10))  # Width x Height in inches
sns.lineplot(x="Frame", y="RMSD", data=rms, linewidth=0.5)
```
